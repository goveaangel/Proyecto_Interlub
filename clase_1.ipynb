```python
data = data.dropna(axis=1, thresh=6)
```

```python
# Función para convertir strings a número (soporta rangos y comas)
def convertir_num(x):
    try:
        if pd.isna(x):
            return np.nan
        if isinstance(x, str):
            # Limpia espacios y reemplaza comas por puntos
            x = x.strip().replace(',', '.')
            # Si tiene un rango tipo "265 - 295", promedia
            if '-' in x:
                a, b = x.replace(' ', '').split('-')
                return (float(a) + float(b)) / 2
            # Si es número directo
            return float(x)
        return x
    except:
        return np.nan

# Detecta columnas con tipo 'object' que parezcan numéricas
cols_convertir = []
for col in data.columns:
    if data[col].dtype == 'object':
        sample = data[col].dropna().astype(str).head(10)
        if sample.str.replace('.', '', 1).str.replace('-', '', 1).str.replace(' ', '').str.isnumeric().any():
            cols_convertir.append(col)

print("Columnas a convertir:", cols_convertir)

# Aplica la conversión
for col in cols_convertir:
    data[col] = data[col].apply(convertir_num)
```

```python
df_categoricas = data.select_dtypes(include=['object'])
df_numericas = data.select_dtypes(include='number')
```
